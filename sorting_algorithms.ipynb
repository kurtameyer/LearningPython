```python
import random
from time import time
import matplotlib.pyplot as plt

```

**1. Merge Sort**

```python
L = [i for i in range(100)]
random.shuffle(L)
print(L)
```

```python

#Slightly different implementation using pop and append. 
def merge(A, B):
  out = [] 
  while(min(len(A), len(B))>0):
    if A[0] > B[0]: 
      to_insert = B.pop(0)  
      out.append(to_insert)
    elif A[0] <= B[0]:
      to_insert = A.pop(0) 
      out.append(to_insert)
  if(len(A)>0):
    for i in A:
      out.append(i)
  if(len(B) > 0):
    for i in B:
      out.append(i)
      
  return(out)
```

```python
def merge_sort(L):
  if len(L)<2:
    return (L[:])
  else:
    mid = len(L)//2
    Left = merge_sort(L[:mid])
    Right = merge_sort(L[mid:])
    return merge(Left, Right)
t1 = time()
output = (merge_sort(L))
t2 = time()
mergesort_time = (t2-t1)*1000
print(output)
print(mergesort_time)
```

**2. Insertion Sort**

```python
L = [i for i in range(100)]
random.shuffle(L)

print(L)
```

```python
def insertion_sort(A):
  for j in range (1, len(A)):
    k = A[j]
    i = j-1
    while i>=0 and A[i]>k:
      A[i+1]=A[i]
      i = i -1
    A[i+1]=k
  return A

print (insertion_sort(L))

t1 = time()
output = (insertion_sort(L))
t2 = time()
insertion_sort_time = (t2-t1)*1000
print(output)


```

**3. Bubble Sort**

```python
L = [i for i in range(100)]
random.shuffle(L)
print(L)
```

```python
def bubble_sort(L):
  n = len(L)

  for i in range(n): # Iterate
    for j in range(0, n-i-1):
      if L[j] > L[j+1]: #Compare
        L[j], L[j+1] = L[j+1], L[j] #Switcharoo
  return L


t1 = time()
output = (bubble_sort(L))
t2 = time()
bubble_sort_time = (t2-t1)*1000
print(output)
print(bubble_sort_time)
```

```python
#Creating lists for graphic below
merges = []
inserts = []
bubbles = []
print("N\tMerge\tInsert\tBubble")
for n in range(100, 5100, 100):
    L = [i for i in range(n)]
    random.shuffle(L)

 
    L_copy = L.copy()
    t1 = time()
    merge_sort(L_copy)
    t2 = time()
    merge_time = (t2 - t1) * 1000
    merges.append(merge_time)

    L_copy = L.copy()
    t1 = time()
    insertion_sort(L_copy)
    t2 = time()
    insert_time = (t2 - t1) * 1000
    inserts.append(insert_time)

  
    L_copy = L.copy()
    t1 = time()
    bubble_sort(L_copy)
    t2 = time()
    bubble_time = (t2 - t1) * 1000
    bubbles.append(bubble_time)

    print(f"{n}\t{merge_time:.1f}\t{insert_time:.1f}\t{bubble_time:.1f}")


    
```

Just for fun: Graphic depicting timing tests for each sorting algorithm. 

```python
plt.style.use('fivethirtyeight')
x_values = list(range(100, 5100, 100))
plt.figure(figsize=(12, 6))
plt.plot(x_values, merges, label='Merge Sort')
plt.plot(x_values, inserts, label='Insertion Sort')
plt.plot(x_values, bubbles, label='Bubble Sort')

plt.xlabel('N')
plt.ylabel('Time in milliseconds')
plt.title('Running Times of Sorting Algorithms')
plt.legend()
plt.show()
```
