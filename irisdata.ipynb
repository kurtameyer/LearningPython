```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
import plotly.express as px
```

Import the dataset and explore and display the features of the dataset.

```python
iris = sns.load_dataset('iris')
```

```python
description = iris.describe()
print(description)
```

```python
iris_df = pd.DataFrame(iris)
iris_df.head(10)
```

Create a scatter plot of the sepal length verses the sepal width.

```python
palette = {'setosa': 'red', 'versicolor': 'orange', 'virginica': 'blue'}
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', palette=palette, data=iris_df)
```

Add the petal length as a third dimension to the plot.

```python
fig = px.scatter_3d(iris_df, x='sepal_length', y='sepal_width', z='petal_length',
                    color='species')
fig.show()
```

Explore the dataset using tools and libraries available in Python.

```python
setosa_summary = iris_df[iris_df['species'] == 'setosa'].describe()
versicolor_summary = iris_df[iris_df['species'] == 'versicolor'].describe()
virginica_summary = iris_df[iris_df['species'] == 'virginica'].describe()
```

```python
species = iris_df['species'].unique()

# Create subplots
fig, axs = plt.subplots(len(species), 2, figsize=(10, 15))

for i, spec in enumerate(species):
    # Sepal Length Histogram
    sns.histplot(data=iris_df[iris_df['species'] == spec], x='sepal_length', kde=True, ax=axs[i, 0])
    axs[i, 0].set_title(f'{spec} - Sepal Length')

    # Sepal Width Histogram
    sns.histplot(data=iris_df[iris_df['species'] == spec], x='sepal_width', kde=True, ax=axs[i, 1])
    axs[i, 1].set_title(f'{spec} - Sepal Width')

plt.tight_layout()
plt.show()
```

```python

custom_palette = ['#FF0000', '#00FF00', '#0000FF']  # Red, Green, Blue
sns.set_palette(custom_palette)
sns.set_style("darkgrid")
sns.set_context("notebook", font_scale=1.2)


sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
```

Calculate the dot product between the sepal length vector vs the sepal width vector.

```python
sepal_length = iris_df['sepal_length'].values
sepal_width = iris_df['sepal_width'].values
dot_product = np.dot(sepal_length, sepal_width)
print("Dot product between sepal length and sepal width:", dot_product)
```

Calculate the norm of the sepal length vectors.

```python
sepal_length = iris_df['sepal_length'].values
norm = np.linalg.norm(sepal_length)
print("Norm of the sepal length vector:", norm)
```

Calculate the distance between the sepal length vector and the sepal width vector.

```python
distance = np.linalg.norm(sepal_length - sepal_width)

# Print the result
print("Distance between sepal length and sepal width vectors:", distance)
```
